Q1) A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.

Based on the question we can understand it is a 2 sample test So use the scripting in that format.

**Ho(NULL HYPOTHESIS) :F&B manager wants to determine whether there is no significant difference in the diameter of the cutlet between two units.
 Ha(ALTERNATE HYPOTHESIS) :F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units.

```python
import pandas as pd
from scipy import stats
import numpy as np
import seaborn as sns
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')
```

```python
unit = pd.read_csv('Cutlets.csv')
```

```python
unit1 = unit['Unit A']
```

```python
unit2 =unit['Unit B']
```

```python
np.mean(unit1)
```

```python
np.mean(unit2)
```

```python
pval = stats.ttest_ind(a=unit1, b= unit2)
```

```python
pval 
```

```python
sns.distplot(unit1)
```

```python
sns.distplot(unit2) 
```

```python
sns.distplot(unit) # both UNIT A AND UNITB 
```

**Here based on the p value we can say that there is no significant difference in the diameter of the cutlets between two units.
And (H0)NULLHYPOTHESIS IS APPLICABLE. P VAL(0.47)> ALPHA VAL(0.05)

Q2) A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.


Here there are 4 laboratories which was greater than 2 samples were we use "ANOVA"

```python
#Ho(NULL HYPOTHESIS): A hospital wants to determine whether there is NO difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list
#Ha(ALTERNATE HYPOTHESIS): A hospital wants to determine whether there is ANY difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list
```

```python
from scipy.stats import norm
```

```python
lab=pd.read_csv('LabTAT.csv')
lab.head()
```
